```python
# loading the dataset
breast_cancer_df = sklearn.datasets.load_breast_cancer()
```

```python
# we need to load it into a pandas dataframe
df = pd.DataFrame(data=breast_cancer_df.data, columns=breast_cancer_df.feature_names)
df.head(5)
```

```python
# adding the target column to the dataframe
df['target'] = breast_cancer_df.target
df.head(5)
```

```python
# splitting the dataset into training and testing sets
X = df.drop('target', axis=1)
y = df['target']
```

```python
X = X.values
y = y.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

## Building Neural Network

```python
X.shape[1]
```

```python
# importing pytorch
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
# setting the seed for reproducibility
torch.manual_seed(41)
```

```python
# create a model class
class Model(nn.Module):
    # input layer (30 features)
    #  hidden layer1 (64 neurons)
    #  hidden layer2 (32 neurons)
    #  output layer (2 neurons)
    def __init__(self, input_features=30, h1=64, h2=32, output_features=2):
        super(Model, self).__init__()
        self.fc1 = nn.Linear(input_features, h1)
        self.fc2 = nn.Linear(h1, h2)
        self.out = nn.Linear(h2, output_features)
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.out(x)
        return x
```

```python
model = Model()
```

```python
%matplotlib inline
```

```python
# converting X features into float tensors
X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)
```

```python
# converting y features into long tensors
y_train = torch.LongTensor(y_train)
y_test = torch.LongTensor(y_test)
```

```python
# setting the criterion of the model
criterion = nn.CrossEntropyLoss()

# setting the optimizer of the model
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# training the model
epochs = 500
losses = []
for i in range(epochs):
    y_pred = model.forward(X_train)
    loss = criterion(y_pred, y_train)
    losses.append(loss)
    print(f'epoch: {i} loss: {loss.item()}')
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
# evaluating the model
with torch.no_grad():
    y_eval = model.forward(X_test)
    loss = criterion(y_eval, y_test)
    print(f'loss: {loss.item()}')
    correct = (torch.argmax(y_eval, dim=1) == y_test).sum().item()
    print(f'accuracy: {correct/len(y_test)}')
```

```python
# saving the model
torch.save(model.state_dict(), 'model.pth')
```

## Cross Validation

```python
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

# K-Fold Cross Validation
kfold = KFold(n_splits=5, shuffle=True, random_state=42)  # 5 folds, shuffle data

# Track performance
fold_accuracies = []
fold_losses = []

for fold, (train_idx, val_idx) in enumerate(kfold.split(X, y)):
    print(f"Training fold {fold+1}/{kfold.get_n_splits()}...")

    # Split the data into train and validation sets for this fold
    X_train, X_val = torch.FloatTensor(X[train_idx]), torch.FloatTensor(X[val_idx])
    y_train, y_val = torch.LongTensor(y[train_idx]), torch.LongTensor(y[val_idx])

    # Initialize the model and optimizer for each fold
    model = Model(input_features=X_train.shape[1])
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    # Train the model
    epochs = 500  # Define the number of epochs for training
    for epoch in range(epochs):
        model.train()  # Set the model to training mode
        optimizer.zero_grad()

        # Forward pass
        y_pred = model(X_train)

        # Compute the loss
        loss = criterion(y_pred, y_train)

        # Backward pass and optimization
        loss.backward()
        optimizer.step()

        if (epoch+1) % 20 == 0:  # Print every 20 epochs
            print(f"Fold {fold+1} - Epoch {epoch+1}/{epochs}, Loss: {loss.item()}")

    # Evaluate the model on the validation set
    model.eval()  # Set the model to evaluation mode
    with torch.no_grad():
        y_val_pred = model(X_val)
        val_loss = criterion(y_val_pred, y_val)

        # Calculate accuracy
        _, predicted_labels = torch.max(y_val_pred, 1)
        accuracy = accuracy_score(y_val.numpy(), predicted_labels.numpy())

    # Track performance for this fold
    fold_accuracies.append(accuracy)
    fold_losses.append(val_loss.item())

    print(f"Fold {fold+1} - Validation Loss: {val_loss.item()}, Accuracy: {accuracy}")

# Calculate average performance across all folds
avg_accuracy = sum(fold_accuracies) / len(fold_accuracies)
avg_loss = sum(fold_losses) / len(fold_losses)

print(f"\nAverage Accuracy: {avg_accuracy}")
print(f"Average Validation Loss: {avg_loss}")
```
